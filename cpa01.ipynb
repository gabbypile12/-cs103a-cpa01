```python
#load the data into a pandas dataframe

df = pd.read_csv('coronavirus_dataset.csv')
data = [row for row in df]
df.head(100)
```

```python
#use pd.describe() to get a rough overview of the data
df.describe()
```

```python
#print the array of columns and the index array
print(df.columns)
```

```python
#create some simple plot of part of the data
x = df['Lat']
y = df['cases']
plt.plot(x,y)
plt.xlabel("Latitude")
plt.ylabel("cases")
plt.title("Cases of COVID-19 in the year 2020")
plt.legend()
#plt.show()
```

```python
#create a pivot table and plot some data from that pivot table
pivot = pd.pivot_table(df, index=['Country.Region'], values=['cases'], columns=['date'])
pivot
```

```python
#use the groupby feature
df.groupby(['Country.Region','date'], as_index=False)['cases'].min()
```
